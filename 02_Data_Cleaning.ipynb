```python
import pandas as pd

df = pd.read_csv(r"C:\Users\pc\OneDrive\Desktop\ecommerce-sales-analytics\data\raw\superstore_data.csv")

print("Raw Data Loaded!")
print("Shape:", df.shape)
```

```python
# Remove spaces, lowercase, replace spaces with underscore
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

print("Column names cleaned!")
print(df.columns.tolist())
```

```python
# Fix "/" instead of "." problem
df['sales']    = df['sales'].astype(str).str.strip().str.replace('/', '.').astype(float)
df['profit']   = df['profit'].astype(str).str.strip().str.replace('/', '.').astype(float)
df['discount'] = df['discount'].astype(str).str.strip().str.replace('/', '.').astype(float)

print("Sales, Profit, Discount fixed!")
print(df[['sales','profit','discount']].head())
```

```python
df['order_date'] = pd.to_datetime(df['order_date'].str.strip())
df['ship_date']  = pd.to_datetime(df['ship_date'].str.strip())

# Extract useful columns
df['month']      = df['order_date'].dt.month
df['year']       = df['order_date'].dt.year
df['month_year'] = df['order_date'].dt.to_period('M')

print("Dates fixed!")
print(df[['order_date', 'ship_date', 'month', 'year', 'month_year']].head())
```

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]

print(f"Duplicates removed: {before - after} rows dropped")
print(f"Final shape: {df.shape}")
```

```python
print("Missing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
```

```python
df.to_csv(r"C:\Users\pc\OneDrive\Desktop\ecommerce-sales-analytics\data\processed\clean_superstore.csv", index=False)

print("Clean data saved to processed folder!")
print("Location: data/processed/clean_superstore.csv")
```
